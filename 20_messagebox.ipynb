# Messagebox
> ### `showinfo`
Is used where we need to show some relevant information to the user.

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.showinfo("information","Information")

top.mainloop()
```

> ### `showwarning`

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.showwarning("warning","Warning")

top.mainloop()
```

> ### `showerror`

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.showerror("error","Error")

top.mainloop()
```

> ### `askquestion`
Is used to ask some question to the user which can be answered in yes or no.

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.askquestion("Confirm","Are you sure?")

top.mainloop()
```

> ### `askokcancel`
Is used to confirm the user's action regarding some application activity.

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.askokcancel("Redirect","Redirecting you to www.javatpoint.com")

top.mainloop()
```

> ### `askyesno`
Is used to ask the user about some action to which, the user can answer in yes or no.

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.askyesno("Application","Got It?")

top.mainloop()
```

> ### `askretrycancel`
Is used to ask the user about doing a particular task again or not.

```python
from tkinter import *
from tkinter import messagebox
top = Tk()
top.geometry("100x100")

messagebox.askretrycancel("Application","try again?")

top.mainloop()
```
